```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import torchvision.models as models
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
# DATASET (CIFAR10 example)

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

num_classes = 10
```

```python
def set_finetune_strategy(model, strategy):
    # Freeze all layers first
    for param in model.parameters():
        param.requires_grad = False

    if strategy == "classifier":
        # Only fc layer trainable
        for param in model.fc.parameters():
            param.requires_grad = True

    elif strategy == "last_block":
        # Unfreeze layer4 + fc
        for param in model.layer4.parameters():
            param.requires_grad = True
        for param in model.fc.parameters():
            param.requires_grad = True

    elif strategy == "full":
        # Unfreeze all
        for param in model.parameters():
            param.requires_grad = True
```

```python
def train_model(model, trainloader, testloader, epochs, optimizer, criterion):
    train_loss_list, test_loss_list = [], []
    train_acc_list, test_acc_list = [], []
    
    epochs = 5
    for epoch in range(epochs):
        model.train()
        running_loss, correct, total = 0, 0, 0

        for images, labels in trainloader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, pred = outputs.max(1)
            correct += pred.eq(labels).sum().item()
            total += labels.size(0)

        train_loss = running_loss / len(trainloader)
        train_acc = 100. * correct / total

        # evaluation
        model.eval()
        test_loss, correct, total = 0, 0, 0

        with torch.no_grad():
            for images, labels in testloader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                test_loss += loss.item()
                _, pred = outputs.max(1)
                correct += pred.eq(labels).sum().item()
                total += labels.size(0)

        test_loss /= len(testloader)
        test_acc = 100. * correct / total

        # Store values
        train_loss_list.append(train_loss)
        test_loss_list.append(test_loss)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)

        print(f"Epoch {epoch+1}: Train Acc={train_acc:.2f}, Test Acc={test_acc:.2f}")

    return train_loss_list, test_loss_list, train_acc_list, test_acc_list
```

```python
import torchvision.models as models
import torch.optim as optim

results = {}
epochs = 5
strategies = ["classifier", "last_block", "full"]

for strat in strategies:
    print(f"\n=== Training: {strat} ===")

    model = models.resnet18(pretrained=True)
    model.fc = nn.Linear(model.fc.in_features, num_classes)
    model = model.to(device)

    set_finetune_strategy(model, strat)

    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    train_loss, test_loss, train_acc, test_acc = train_model(
        model, trainloader, testloader, epochs, optimizer, criterion
    )

    # Save checkpoint
    torch.save(model.state_dict(), f"resnet18_{strat}.pth")

    # Store results
    results[strat] = {
        "train_loss": train_loss,
        "test_loss": test_loss,
        "train_acc": train_acc,
        "test_acc": test_acc
    }
```

```python
import matplotlib.pyplot as plt

def plot_results(results):
    # Accuracy Plot
    plt.figure()
    for strat in results:
        plt.plot(results[strat]["test_acc"], label=strat)
    plt.title("Test Accuracy Comparison")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.savefig("accuracy_comparison.png")
    plt.show()

    # Loss Plot
    plt.figure()
    for strat in results:
        plt.plot(results[strat]["test_loss"], label=strat)
    plt.title("Test Loss Comparison")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.savefig("loss_comparison.png")
    plt.show()

plot_results(results)
```

```python
print("\nFinal Accuracy Comparison:")
for strat in results:
    print(f"{strat}: {results[strat]['test_acc'][-1]:.2f}%")
```
